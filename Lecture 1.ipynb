## What is QSAR?

QSAR stands for **Quantitative Structure–Activity Relationship**.

In simple words:
> QSAR tries to learn the relationship between a molecule’s **chemical structure**
and its **biological activity** using data and mathematics.

Instead of experimentally testing thousands of compounds in the lab,
QSAR allows us to:
- Test a smaller number of compounds experimentally
- Learn patterns from this data
- Predict activity of new compounds computationally

This approach saves **time, cost, and experimental effort** in drug discovery.



## Introduction to the ChEMBL Database

ChEMBL is a **manually curated database** maintained by EMBL-EBI.
It contains:

- Chemical structures of small molecules
- Biological targets such as enzymes and receptors
- Experimentally measured bioactivity data such as IC50, EC50, Ki, and Kd


**Experimentally measured bioactivity data** are numbers obtained from lab experiments that tell us **how strongly a molecule (drug/compound) interacts with a biological target** (like a protein or enzyme).

### Common types:

### **1. IC₅₀ (Inhibitory Concentration 50%)**

* The concentration of a compound needed to **reduce the activity of a protein by 50%**.
* **Lower IC₅₀ = stronger inhibitor**
* Example:
  If Drug A has IC₅₀ = 5 µM and Drug B has IC₅₀ = 50 µM → **Drug A is more potent**

---

### **2. EC₅₀ (Effective Concentration 50%)**

* The concentration of a compound needed to produce **50% of its maximum biological effect**.
* Used when a compound **activates** a receptor.
* **Lower EC₅₀ = stronger effect**

---

### **3. Ki (Inhibition Constant)**

* Measures how tightly an inhibitor binds to an enzyme.
* **Lower Ki = stronger binding**
* More “pure” binding measure, less dependent on experimental conditions.

---

### **4. Kd (Dissociation Constant)**

* Measures how tightly **any ligand binds to a target**.
* **Lower Kd = tighter binding**
* Example:
  Low Kd means the ligand **does not easily fall off** the protein.

---

### **In one line (easy to remember):**

* **IC₅₀ / EC₅₀ → how much drug is needed**
* **Ki / Kd → how strongly the drug binds**
* **Lower value = better/stronger interaction**



Why ChEMBL is important for QSAR:
- Data is derived from **wet-lab experiments**
- Targets are well annotated
- Units are standardized

In QSAR modeling, ChEMBL is one of the **most trusted public data sources**.


## Step 2: Importing Python Libraries

Now we import the required libraries:

- `pandas`: used to handle tabular data (similar to Excel)
- `chembl_webresource_client`: used to access ChEMBL data


```python

import pandas as pd
from chembl_webresource_client.new_client import new_client
```


## Step 3: Selecting a Biological Target

In QSAR, bioactivity data is always associated with a **specific biological target**.

In this bootcamp, we use **Human Acetylcholinesterase**:
- An important enzyme in neurotransmission
- A common drug target for Alzheimer’s disease

ChEMBL ID for this target: **CHEMBL220**


```python

target = new_client.target
targets = pd.DataFrame.from_dict(target.search('human acetylcholinesterase'))
targets[['target_chembl_id', 'pref_name', 'organism']]
```


## Step 4: Understanding IC50 Bioactivity Data

IC50 stands for **Inhibitory Concentration 50%**.

It is the concentration of a compound required to inhibit
50% of the biological activity of a target.

Key points:
- Lower IC50 → stronger inhibitor
- Higher IC50 → weaker inhibitor
- IC50 values are usually measured in nanomolar (nM)

IC50 is one of the most commonly used endpoints in QSAR modeling.



## Step 5: Retrieving IC50 Data from ChEMBL

Now we download **experimental IC50 data** for our selected target.

We explicitly filter:
- Target = Acetylcholinesterase
- Bioactivity type = IC50


```python
activity = new_client.activity
res = activity.filter(
    target_chembl_id='CHEMBL220',
    standard_type='IC50'
)

df = pd.DataFrame.from_dict(res)
df.head()
```


## Step 6: Initial Exploration of the Dataset

Before cleaning the data, it is important to understand:
- How many data points we have
- The range of IC50 values

This gives us a sense of dataset quality.


```python

print("Total records:", df.shape[0])
df['standard_value'].describe()
```


## Step 7: Data Cleaning – Why It Is Necessary

Machine learning models are **very sensitive** to bad data.

We must remove:
- Compounds with missing IC50 values
- Compounds with missing chemical structures (SMILES)
- Duplicate molecules

Clean data leads to reliable QSAR models.


```python

df = df[df.standard_value.notna()]
df = df[df.canonical_smiles.notna()]
df = df.drop_duplicates(['canonical_smiles'])

print("Records after cleaning:", df.shape[0])
```


## Step 8: Labeling Compounds Based on Activity

For classification-based QSAR models,
we convert continuous IC50 values into **activity classes**.

Thresholds used:
- Active: IC50 ≤ 1000 nM
- Inactive: IC50 ≥ 10000 nM
- Intermediate: values in between

These thresholds are biologically motivated and commonly used.


```python

def label_activity(x):
    if x <= 1000:
        return 'active'
    elif x >= 10000:
        return 'inactive'
    else:
        return 'intermediate'

df['class'] = df.standard_value.astype(float).apply(label_activity)
df[['molecule_chembl_id', 'standard_value', 'class']].head()
```


## Step 9: Saving the Curated QSAR Dataset

Finally, we save the cleaned and labeled dataset.

This dataset will be used in the next lecture
to calculate molecular descriptors and build QSAR models.


```python

df_final = df[['molecule_chembl_id', 'canonical_smiles', 'standard_value', 'class']]
df_final.to_csv('bioactivity_curated.csv', index=False)
df_final.head()
```
